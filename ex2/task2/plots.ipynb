```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from IPython.display import display
```

```python
HEADER = ["tests_done", "N", "total_time", "time_per_test", "max_runtime", "min_runtime"]
HEADER2 = ["grid_sizes", "block_size"] + HEADER
def csv_parse(path, header=HEADER):
    if header:
        df = pd.read_csv(path, sep=";", names=header)
    else:
        df = pd.read_csv(path, sep=";")
    #display(df.head())
    return df
```

# Ex2.2. a)

```python
df = csv_parse("/Users/peterholzner/Code/many_cores/ex2/task2/out/data1.csv", header=False)
display(df)
df2 = csv_parse("/Users/peterholzner/Code/many_cores/ex2/task2/out/data2.csv", header=False)
display(df2)
df3 = csv_parse("/Users/peterholzner/Code/many_cores/ex2/task2/out/data3.csv", header=False)
display(df3)


vec = df["N"]
fig, ax  = plt.subplots(figsize=(10,7));
df.plot(x="N", y="time_per_test", ax=ax, marker="o", xticks=vec, logx=True, logy=True);
df2.plot(x="N", y="time_per_test", ax=ax, marker="o", xticks=vec, logx=True, logy=True);
df3.plot(x="N", y="time_per_test", ax=ax, marker="o", xticks=vec, logx=True, logy=True);
ax.set_title("Runtimes of dot-products (block_size = 256)", fontsize=18);
ax.set_ylabel("runtime [s]");
ax.legend(["2 CUDA kernels", "CUDA + CPU","atomic CUDA"]);
ax.grid();
fig.savefig("/Users/peterholzner/Code/many_cores/ex2/task2/out/fig_ex2_2.pdf")
```
